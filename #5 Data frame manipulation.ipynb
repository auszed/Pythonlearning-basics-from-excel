# Selecting One Column from a Dataframe

```python
# import pandas
import pandas as pd
# import numpy
import numpy as np
```

```python
# reading the csv file
df_exams = pd.read_csv('hotel_bookings.csv')
df_exams.head()
```

# Selecting one column [ ]

using series in 1d and the size its immutable but the data its mutable

### Syntax 1

```python
# select a column with [] (preferred way to select a column)
df_exams['customer_type']
```

```python
# check out the data type of a column
type(df_exams['customer_type'])
```

```python
# series: attributes and methods
df_exams['customer_type'].index
```

```python
df_exams['customer_type'].head()
```

### Syntax 2

this syntax its only valid for one value or a serie

```python
# select a column with .
df_exams.customer_type
```

if there its an space in the title we need to use the other way so we dont get an error

```python
# select a column with . (pitfalls)
# this will fail if there its a space in the title of the column
df_exams.math score
```

## Selecting Two or More Columns from a Dataframe [[ ]]

using dataframe in 2d and the size its mutable but the data its mutable also we need to use double [[ ]]


```python
# select 2 or more columns using [[]]
df_exams[['customer_type', 'deposit_type', 'arrival_date_year']]
```

```python
# check out the data type of the selection
type(df_exams[['customer_type', 'deposit_type','arrival_date_year']])
```

```python
# we can't select 2 or more columns with the "."
df_exams.'customer_type', 'deposit_type'
```

# Add New Column to a Dataframe

### Adding a new column with a scalar value

all the rows have the same value with this form

```python
# adding new column to dataframe with a unique value

df_exams['language score'] = 70
df_exams
```

```python
df_exams.shape
```

### Adding a new column with an array

The dataframe has 119390 rows, so we need to create an array with 1000 elements also
the number of rows have to mach the number of rows

```python
# create an array of 119390 elements
# we create this array
Language_array = np.arange(0, 119390)
```

```python
# length of the array
len(Language_array)
```

```python
# adding a new column to dataframe with an array
df_exams['language score'] = Language_array
df_exams
```

### creating random reaÃ± numbers with out decimals

```python
# create random integer numbers between 1 and 100
int_language_score = np.random.randint(1, 100, size=119390)
```

```python
# min value inclusive and high value exclusive
print(min(int_language_score))
print(max(int_language_score))
```

### change a value of a whole cell

```python
# adding a new column to dataframe with an array
df_exams['language score'] = int_language_score
df_exams
```
