# pivot_table()

import an excel file

```python
# read and show the dataset
df_sales = pd.read_excel('Online Direct Marketer.xlsx')
df_sales
```

summarize all the table ussing the gender as filter with a sum agregation also only numerical data because we use the sum

```python
# make a pivot table and add an aggregate function
df_sales.pivot_table(index='Gender',
                     aggfunc='sum').round(1)
```

filter by only the columns that we want

```python
# make a pivot table, add an aggregate function and select some columns 
df_sales.pivot_table(index='Gender',
                     values=['Salary', 'AmountSpent'],
                     aggfunc='sum').round(1)
```

In the colum value we only cvan add 1 variable i already try 2 but it was not possible.

```python
# make a pivot tables that says how much male and female spend in each category
df_sales.pivot_table(index='Gender',
                     columns='Age',
                     values='AmountSpent',
                     aggfunc='sum').round(1)
```

##### Data Visualization with Panda

```python
df_pop_raw = pd.read_csv('population_total.csv')
```

```python
df_pop_raw.head(3)
```

```python
df_pop_raw.dropna(inplace=True)       
```

```python
df_pop = df_pop_raw.pivot(index = 'year', columns = 'country', values='population') 
```

```python
df_pop = df_pop[['United States', 'India', 'China', 
                          'Indonesia', 'Brazil']]
```

# Lineplot

kind = the kind of plot we want can be bar , line , etc

figsize = change the size of the graph  (x, y)

```python
df_pop.plot(kind='line',
              xlabel='Years', ylabel='country',
              title='gdppc (1950-2020)',
              figsize=(15,4))
```

# Barplot

### Single Barplot

allows to see the index in this dataframe

```python
df_pop.index
```

select and create a new table with the row that we already select

```python
# selecting only one year (2020)
df_pivot_2020 = df_pop[df_pop.index.isin([2020])]
```

```python
# showing new dataframe
df_pivot_2020
```

we transpose so this means that the rows are swap with the columns so its important to make the swap

```python
# transpose dataframe from rows to columns or viceversa
df_pivot_2020 = df_pivot_2020.T
```

```python
# making barplot
df_pivot_2020.plot(kind='bar', color='orange',
                   xlabel='Years', ylabel='Population',
                   title='Population in 2020')
```

### Barplot grouped by “n” variables

select several years for the bar plot

```python
# selecting a few years
df_pivot_sample = df_pop[df_pop.index.isin([1980, 1990, 2000, 2010, 2020])]
df_pivot_sample
```

```python
# making grouped barplot
df_pivot_sample.plot(kind='bar',
                     xlabel='Years', ylabel='Population',
                     figsize=(8,4))
```

# Piechart

rename => allows to rename a row, but we are trying to convert a value from a number to a string '2000' also with
inplace=True => we overwite the dataframe

```python
# changing column name 
df_pivot_2020.rename(columns={2020:'2020'}, inplace=True)
```

```python
# making the plot
df_pivot_2020.plot(kind='pie',
                   y='2020',
                   title='Population in 2020 (%)' )
```
