# data import

```python
# import pandas
import pandas as pd
```

```python
# reading the csv file
df_exams = pd.read_csv('sales.csv')
df_exams.head()
```

```python
df_exams.dtypes
```

# Sort a Dataframe # data

```python
# sort by one column
df_exams.sort_values('qty_ordered').head()
```

```python
# sort descending by one column
df_exams.sort_values('qty_ordered', ascending=False).head()
```

### sort by multiple columns [ ]

remember to add [ ] because we are using multiple values

```python
# sort descending by multiple columns
df_exams.sort_values(['order_date', 'qty_ordered'], ascending=False).head()
```

```python
df_exams.head(3)
```

### inplace allows to change the metrics in the whole data

as you can see above this was the value before the update, but when we used inplace the whole dataframe its update

```python
# sort descending by multiple columns and update dataframe
df_exams.sort_values(['order_date', 'qty_ordered'], ascending=False, inplace=True)
```

```python
df_exams.head(3)
```

# Sort a Dataframe string or text data

```python
# sort descending with a key function
df_exams.sort_values(['County'], ascending=True, key=lambda a:a.str.lower()).head(3)
```

```python
df_exams.sort_values(['County'], ascending=True).head(3)
```
